```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('./acceleration_2024-08-10_12-24-13.csv')
df2 = pd.read_csv('./acceleration_2024-08-10_13-17-29.csv')


fig = px.line(df, x = 'time', y = ['ax', 'ay', 'az', 'atotal'], title='acceleration', labels={'value': 'Acceleration', 'variable': 'Axis', 'time_column_name': 'Time'})
fig2 = px.line(df2, x = 'time', y = ['ax', 'ay', 'az', 'atotal'], title='acceleration', labels={'value': 'Acceleration', 'variable': 'Axis', 'time_column_name': 'Time'})

fig.show()
fig2.show()
```
